```python
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = None  # Generated automatically
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

class Menu:
    def __init__(self):
        self.food_items = []
        self.food_id_counter = 1

    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        food_item.food_id = self.food_id_counter
        self.food_id_counter += 1
        self.food_items.append(food_item)

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for item in self.food_items:
            if item.food_id == food_id:
                item.name = name
                item.quantity = quantity
                item.price = price
                item.discount = discount
                item.stock = stock
                break

    def view_food_items(self):
        print("FoodID\tName\tQuantity\tPrice\tDiscount\tStock")
        for item in self.food_items:
            print(f"{item.food_id}\t{item.name}\t{item.quantity}\t{item.price}\t{item.discount}\t{item.stock}")

    def remove_food_item(self, food_id):
        for item in self.food_items:
            if item.food_id == food_id:
                self.food_items.remove(item)
                break

menu = Menu()
menu.add_food_item("Burger", "1 piece", 5.99, 0.2, 50)
menu.add_food_item("Pizza", "8 slices", 12.99, 0.15, 30)

print("Initial Menu:")
menu.view_food_items()

menu.edit_food_item(1, "Veggie Burger", "1 piece", 6.99, 0.25, 40)
menu.remove_food_item(2)

print("\nUpdated Menu:")
menu.view_food_items()
```

```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.user_id = None  # Generated automatically
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.order_history = []

class FoodItem:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.price = price

class Restaurant:
    def __init__(self):
        self.users = []
        self.food_menu = [
            FoodItem("Tandoori Chicken (4 pieces)", "4 pieces", 240),
            FoodItem("Vegan Burger (1 Piece)", "1 piece", 320),
            FoodItem("Truffle Cake (500gm)", "500gm", 900)
        ]

    def register_user(self, full_name, phone_number, email, address, password):
        user = User(full_name, phone_number, email, address, password)
        user.user_id = len(self.users) + 1
        self.users.append(user)
        return user.user_id

    def login_user(self, email, password):
        for user in self.users:
            if user.email == email and user.password == password:
                return user
        return None

    def place_order(self, user, selected_items):
        order_details = []
        total_price = 0
        for item_index in selected_items:
            if 1 <= item_index <= len(self.food_menu):
                food_item = self.food_menu[item_index - 1]
                order_details.append(food_item)
                total_price += food_item.price

        if order_details:
            user.order_history.append(order_details)
            return total_price
        else:
            return 0

    def view_order_history(self, user):
        print("Order History for User:", user.full_name)
        for i, order in enumerate(user.order_history, start=1):
            print(f"Order {i}:")
            for item in order:
                print(f"- {item.name} [{item.price} INR]")
            print()

    def update_user_profile(self, user, full_name, phone_number, address, password):
        user.full_name = full_name
        user.phone_number = phone_number
        user.address = address
        user.password = password


restaurant = Restaurant()
user_id = restaurant.register_user("John Doe", "1234567890", "john@example.com", "123 Main St", "password123")


logged_in_user = restaurant.login_user("john@example.com", "password123")
if logged_in_user:
    print("Logged in as:", logged_in_user.full_name)


selected_items = [2, 3]
total_price = restaurant.place_order(logged_in_user, selected_items)
if total_price > 0:
    print("Order Details:")
    for item_index in selected_items:
        if 1 <= item_index <= len(restaurant.food_menu):
            food_item = restaurant.food_menu[item_index - 1]
            print(f"- {food_item.name} [{food_item.price} INR]")
    print("Total Price:", total_price, "INR")


restaurant.view_order_history(logged_in_user)


restaurant.update_user_profile(logged_in_user, "John Smith", "9876543210", "456 Elm St", "newpassword456")
print("Updated Profile:")
print("Full Name:", logged_in_user.full_name)
print("Phone Number:", logged_in_user.phone_number)
print("Address:", logged_in_user.address)
```
